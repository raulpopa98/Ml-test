```python
#!wget -O questions_classification_data.csv 'https://drive.google.com/file/d/1oFOo-JAuIixowXbRMLPYcX5-w10aOgQl/view?usp=sharing'
DF = pd.read_csv('questions_classification_data.csv')  # reading the housepricedata.csv file
DF= DF.fillna('no tags')
DF # show the data set as csv
```

```python
DataFrame = DF.values # convert data into arrays for further processing
DataFrame
```

```python
X = DataFrame[:, 1:2]  # input data set
# From DataFrame take the second row
Y = DataFrame[:, 1:2]  # input data set
print ('X = \n\n', X)
print ('Y = \n\n', Y)
```

```python
from sklearn.model_selection import train_test_split
# val_and_test data set size will be 30% of the overall dataset, while train set will be 70%
X_train, X_val_and_test, Y_train, Y_val_and_test=train_test_split(X, Y, test_size=0.3)
# split the val_and_test subset: 
# - test set 60% (meaning 60% * 30% = 18% from the full dataset)
# - validation set 40% (meaning 12% from the full data set)
X_val, X_test, Y_val, Y_test= train_test_split(X_val_and_test, Y_val_and_test, test_size=0.6)
X_train, Y_train= np.array(X_train), np.array(Y_train)
print('Size of training set  -  X_train: ', X_train.shape, ';   Y_train: ', Y_train.shape)
print('Size of validation set  -  X_val: ', X_val.shape, ';   Y_val: ', Y_val.shape)
print('Size of independent test set  -  X_test: ', X_test.shape, ';   Y_test: ', Y_test.shape)
```

```python
#TensorFlow and Keras
import tensorflow as tf  
from tensorflow import keras 
from keras.models import Sequential # import the Sequential model
from keras.layers import Dense # import the "Dense" - type layer
from keras import layers
```

```python
#Specify the architecture with the Keras Sequential model.
model = Sequential([ # we will store our model in the variable ‘model’, and we’ll describe it sequentially (layer by layer) in between the square brackets
                    Dense(16, activation='relu', input_shape=(1,), name = 'HiddenLayer1'), # we have our first layer as a dense layer with 16 neurons, ReLU activation and the input shape is 1 since we have 1 input features
                    Dense(8, activation='relu', name = 'HiddenLayer2'), # our second layer is also a dense layer with 8 neurons, ReLU activation; not necessary to specify the input shape.
                    Dense(1, activation='sigmoid', name = 'OutputLayer'), # output layer is a dense layer with 1 neuron, sigmoid activation.
    ])
```

```python
#Disply the model
print("The generated Sequential model is:")
model.summary()
import os
print("\nCurrent Working Directory where the model will be saved is:" , os.getcwd())
```

```python
#Compile the model for training (optimizer, loss function, metrics)
model.compile(optimizer='adam', 
              loss='categorical_crossentropy',
              metrics=['accuracy']) # we want to track accuracy on top of the loss function
```

```python
# Train the model
hist = model.fit(
    X_train,   # input data (features)
    Y_train,   # target data
    batch_size = 32,  # number of samples per gradient update (default 32), size of mini-batch.
    epochs=100,  #  number of epochs to train the model
                 # An epoch is an iteration over the entire x and y data provided
    verbose = 2, # verbosity mode (0 - silent, 1 - progress bar, 2 - one line per epoch)
    validation_data=(X_val, Y_val)) # Data on which to evaluate the loss and any model metrics 
                                    # at the end of each epoch. The model will not be trained on this data
```

```python
#Evaluate and use the ANN model
import matplotlib.pyplot as plt # imports
# plot loss values for training set and for validation set
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13,5))  # 1 row; 2 cols
fig.suptitle('Loss and accuracy during training')

ax1.plot(hist.history['loss']) # extract the loss values from the history object generated in the training phase
ax1.plot(hist.history['val_loss'])
ax1.set_title('Model loss during training') 
ax1.set_ylabel('Loss')
ax1.set_xlabel('Epoch') 
ax1.legend(['Train', 'Validation'], loc='upper right') 

ax2.plot(hist.history['accuracy'])
ax2.plot(hist.history['val_accuracy'])
ax2.set_title('Model accuracy')
ax2.set_ylabel('Accuracy')
ax2.set_xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='lower right')
plt.show()
```

```python
#Evaluate the ANN model on the independent test set
test_loss, test_accuracy = model.evaluate(
    X_test, 
    Y_test,
    verbose = 2)
print('\ntest_loss = ', test_loss)
print('test_accuracy = ', test_accuracy)
```
